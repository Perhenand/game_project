```python
def prep_game():
    """Ask user for list of players"""
    pl = ""
    while not pl:
        pl = input("Input player names, separate names by comma: ")
    return [i.strip() for i in pl.split(",")]
```

```python
def replay():
    return input("\n\nDo you wanna play again? Enter Yes or No: ").lower().startswith("y")
```

```python
def podplace(res):
    res = sorted(res,key=lambda e: e[1])
    res = sorted(res,key=lambda e: e[2], reverse=True)
    
    if res[0][2]:
        print(f"{res[0][0]} is the winner with a time of {res[0][1]}\n\n")
        print("PLAYER\t\t\tFINISHED\tTIME")
        print("-----------------------------------------------")
        for p,t,st in res:
            #print(f"{p}\t\t\t{st}\t\t{t if st else ''}")
            print(f"{p+' '*(24-len(p))}{str(st)+' '*(16-len(str(st)))}{t if st else ''}")
    else:
        print("No one finished the race!\n")
        print("All players feel overwhelmed by shame!")
```

```python
def upresult(res,pl,time,st):
    res.append((pl,time,st))
    return res
```

```python
def pathgen():
    bana = []
    bana.append((-10,-50,50))
    bana.append((0,-50,50))
    bana.append((100,-55,45))
    bana.append((200,-63,37))
    bana.append((250,-70,30))
    bana.append((300,-93,7))
    bana.append((350,-115,-15))
    bana.append((400,-145,-45))
    bana.append((450,-180,-80))
    bana.append((500,-240,-140))
    bana.append((550,-300,-200))
    bana.append((600,-310,-210))
    bana.append((650,-290,-190))
    bana.append((700,-280,-180))
    bana.append((750,-272,-172))
    bana.append((800,-233,-133))
    bana.append((850,-190,-90))
    bana.append((900,-120,-20))
    bana.append((950,-70,30))
    bana.append((1000,50,150))
    bana.append((1050,130,230))
    bana.append((1100,167,267))
    bana.append((1150,162,262))
    bana.append((1200,156,256))
    bana.append((1250,152,252))
    bana.append((1300,140,240))
    bana.append((1350,103,183))
    bana.append((1400,65,115))
    bana.append((1450,20,60))
    bana.append((1500,17,40))
    bana.append((1550,20,40))
    bana.append((1600,5,35))
    bana.append((1700,-10,30))
    bana.append((1800,-12,25))
    bana.append((1900,-20,20))
    bana.append((2000,-30,30))
    ymax = bana[-1][0]
    return (ymax,bana)
```

```python
def clashcheck(bn,x,y):
    tmp = []
    for i in range(len(bn)):
        if (bn[i][0] == y) or (bn[i][0] < y and i==len(bn)-1):
            a,b,c = bn[i]
            if b<=x<=c:
                return False
        elif bn[i][0] < y and bn[i+1][0] > y:
            a1,b1,c1 = bn[i]
            a2,b2,c2 = bn[i+1]
            b = b1+(b2-b1)*(y-a1)/(a2-a1)
            c = c1+(c2-c1)*(y-a1)/(a2-a1)
            if b<=x<=c:
                return False
    return True
```

```python
def rungame(pl,bana,ymax):
    x, y, i = 0, 0, 0
    gimal = False
    tid = 0
    pt1 = [(xs,yg) for yg,xs,xe in bana]
    pt2 = [(xe,yg) for yg,xs,xe in bana]
    
    win = turtle.Screen()
    win.bgcolor("white")
    win.title("SnowRush")
    win.setup(width=1000, height=600)
    win.clear()
    
    # Turtle for the skier
    tt = turtle.Turtle()
    tt.speed(0)
    tt.shape("triangle")
    tt.shapesize(0.5,1,1)
    tt.color("green", "green")
    tt.radians()
    tt.rt(-math.pi/2)
    tt.pu()
    
    # Turtle for the slope edges
    st = turtle.Turtle()
    st.speed(0)
    st.pensize(2)
    st.color("dark grey")
    st.penup()
    st.hideturtle()
    
    # Turtle for the clock
    sc = turtle.Turtle()
    sc.speed(0)
    sc.color("dark grey")
    sc.penup()
    sc.hideturtle()
    sc.goto(240, 260)
    sc.write("Time: ",align="left", font=("Courier", 18, "normal"))
    
    def rlar():
        tmp = tt.heading()
        if tmp<=math.pi-0.15: tt.rt(-0.15)
        
    def rrar():
        tmp = tt.heading()
        if tmp>=0.15: tt.rt(0.15)
        
    def elar():
        tmp = tt.shapesize()[1]
        tt.shapesize(stretch_len=tmp+0.1)
        
    def shar():
        tmp = tt.shapesize()[1]
        tt.shapesize(stretch_len=tmp-0.1 if tmp>1 else 1)

    # Bind keys
    win.listen()
    win.onkey(rlar,'Left')
    win.onkey(rrar,'Right')
    win.onkey(elar,'Up')
    win.onkey(shar,'Down')
    win.onkey(rlar,'a')
    win.onkey(rrar,'d')
    win.onkey(elar,'w')
    win.onkey(shar,'s')
        
    def koord():
        ri = tt.heading()
        fr = (tt.shapesize()[1]-1)/1+1/10
        dx = round(math.cos(ri)*fr,2)
        dy = round(math.sin(ri)*fr,2)
        return (x+dx,y+dy)

    def drwpth(dx,dy):
        win.tracer(0, 0)
        st.clear()
        for m in [pt1,pt2]:
            st.pu()
            for n,v in enumerate(m):
                st.goto(v[0]-dx,v[1]-dy)
                if n == 0:
                    st.pd() 
        win.tracer(1, 0)
        
    def uptid():
        win.tracer(0, 0)
        sc.clear()
        tmp = time.time()-tids
        sc.write(f"Time: {int(tmp//60)}:{round(tmp%60,2)}",align="left", font=("Courier", 24, "normal"))
        win.tracer(1, 0)
        
    def stfst():  
        if clashcheck(bana,xl,yl):
            tmp = time.time()-stid
            tid = f"{int(tmp//60)}:{round(tmp%60,2)}"
            win.bye()
        elif yl >= ymax:
            tmp = time.time()-stid
            tid = f"{int(tmp//60)}:{round(tmp%60,2)}"
            st = True
            win.bye()
    
    tids = time.time()
    while True:
        win.update()
        x,y = koord()
        drwpth(x,y)
        uptid()
        
        if clashcheck(bana,x,y):
            tmp = time.time()-tids
            tid = f"{int(tmp//60)}:{round(tmp%60,2)}"
            win.bye()
            break
        elif y >= ymax:
            tmp = time.time()-tids
            tid = f"{int(tmp//60)}:{round(tmp%60,2)}"
            gimal = True
            win.bye()
            break
    
    try:
        turtle.mainloop()
    except:
        pass
    return (tid,gimal)
```

```python
import turtle
import time
import math
from IPython.display import clear_output

player = prep_game()

if input("Are you ready to play? Enter Yes or No: ").lower().startswith("y"):
    game_on = True
else:
    game_on = False
    
while game_on:
    ymax, bana = pathgen()
    res = []
    tid = 0
    st = False
    for pl in player:
        tid,st = rungame(pl,bana,ymax)
        res = upresult(res,pl,tid,st)
    clear_output()
    podplace(res)
    
    if not replay():
        break
```
